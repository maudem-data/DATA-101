## 1. What is Exploratory Data Analysis (EDA)?

Exploratory Data Analysis (EDA) is a crucial process in the data science workflow that involves examining and summarizing the main characteristics of a dataset, often with visual methods. Its primary goals are:

*   **Understanding the Data**: To gain insights into the structure, patterns, and relationships within the data.
*   **Identifying Anomalies**: To detect outliers, errors, or missing values that might impact subsequent analysis.
*   **Formulating Hypotheses**: To generate hypotheses about the data that can be tested with more formal statistical methods.
*   **Feature Engineering**: To guide the selection of appropriate statistical models and feature engineering techniques.
*   **Preparing for Modeling**: To clean, transform, and prepare the data for machine learning model training.

EDA helps to uncover underlying structures, extract important variables, detect outliers and anomalies, test underlying assumptions, and develop parsimonious models.

## 2. General Process of EDA

The EDA process can be broken down into several logical phases, each contributing to a comprehensive understanding of the dataset. This structured approach helps in systematically uncovering insights, preparing the data for modeling, and making informed decisions.

### Phase 1: Data Loading and Initial Inspection

**Objective**: To get a first look at the data, understand its structure, and identify immediate issues like data types or inconsistent entries.

**Common Techniques/Visualizations**:
*   **Loading Data**: Using libraries like `pandas` to load data from various sources (CSV, Excel, databases, etc.).
*   **`df.head()`, `df.tail()`, `df.sample()`**: Viewing the first, last, or random rows to get a glimpse of the data.
*   **`df.info()`**: Obtaining a summary of the DataFrame, including data types, non-null values, and memory usage.
*   **`df.describe()`**: Generating descriptive statistics (count, mean, std, min, max, quartiles) for numerical columns.
*   **`df.shape`**: Checking the number of rows and columns.
*   **`df.columns`**: Listing all column names to check for relevance or naming inconsistencies.
*   **`df.dtypes`**: Reviewing the data types of each column.

### Phase 2: Univariate Analysis

**Objective**: To understand the distribution and characteristics of individual variables independently.

**Common Techniques/Visualizations**:
*   **For Numerical Variables**:
    *   **Histograms**: To visualize the distribution, central tendency, and spread of data.
    *   **Box Plots**: To identify outliers and understand the distribution's five-number summary (min, max, median, quartiles).
    *   **Density Plots (KDE)**: To show the probability density function of a variable.
    *   **Summary Statistics**: Mean, median, mode, standard deviation, variance, skewness, kurtosis.
*   **For Categorical Variables**:
    *   **Bar Charts/Count Plots**: To display the frequency of each category.
    *   **Pie Charts**: To show the proportion of each category (use sparingly for many categories).
    *   **Frequency Tables**: To list the counts and proportions of each category (`df['column'].value_counts()`).

### Phase 3: Bivariate Analysis

**Objective**: To explore relationships between pairs of variables, identifying potential correlations or dependencies.

**Common Techniques/Visualizations**:
*   **Numerical vs. Numerical**:
    *   **Scatter Plots**: To visualize the relationship between two continuous variables and identify patterns or correlations.
    *   **Correlation Matrix/Heatmap**: To show the correlation coefficients between all pairs of numerical variables.
    *   **Line Plots**: For time-series data or when observing trends.
*   **Numerical vs. Categorical**:
    *   **Box Plots/Violin Plots**: To compare the distribution of a numerical variable across different categories.
    *   **Grouped Bar Charts**: To show the mean or median of a numerical variable for each category.
*   **Categorical vs. Categorical**:
    *   **Stacked Bar Charts/Grouped Bar Charts**: To compare the distribution of one categorical variable across categories of another.
    *   **Crosstabulations (Contingency Tables)**: To summarize the relationship between two or more categorical variables (`pd.crosstab()`).

### Phase 4: Handling Missing Values and Outliers

**Objective**: To identify and address data quality issues that can impact the reliability of analysis and model performance.

**Common Techniques/Visualizations**:
*   **Missing Values**:
    *   **`df.isnull().sum()`**: To count missing values per column.
    *   **Heatmaps of Missing Data**: To visualize patterns of missingness across the dataset.
    *   **Imputation**: Filling missing values using strategies like mean, median, mode, or more advanced methods (e.g., K-Nearest Neighbors imputation).
    *   **Deletion**: Removing rows or columns with too many missing values (use cautiously).
*   **Outliers**:
    *   **Box Plots**: As mentioned in univariate analysis, effective for identifying outliers.
    *   **Z-score or IQR Method**: Statistical methods to detect data points significantly different from the rest.
    *   **Scatter Plots**: Can reveal outliers in bivariate relationships.
    *   **Treatment**: Transformation (e.g., log transform), capping/flooring, or removal (if justified and minimal).

These phases are iterative and often revisited as new insights emerge, forming a comprehensive approach to data understanding and preparation.

## Data Loading and Initial Inspection


Load the chosen dataset (Titanic) and perform initial checks such as viewing the first few rows, checking data types, and summarizing basic statistics.

To begin, I'll import the pandas library and load the 'titanic.csv' dataset form seaborn into a DataFrame, and then perform initial data checks such as displaying the first few rows, checking data types and non-null counts, showing descriptive statistics for numerical columns, and reporting the DataFrame's dimensions.



```python
import pandas as pd
import seaborn as sns

# Load the Titanic dataset using seaborn
df = sns.load_dataset('titanic')

# Display the first 5 rows
print("First 5 rows of the DataFrame:")
print(df.head())

# Print a concise summary of the DataFrame
print("\nDataFrame Info:")
df.info()

# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(df.describe())

# Check the number of rows and columns
print("\nDataFrame Shape (Rows, Columns):")
print(df.shape)
```

## Univariate Analysis

Analyze individual features using descriptive statistics, histograms, box plots, and value counts to understand their distributions and characteristics.

1. Import the necessary libraries for plotting, matplotlib.


```python
import matplotlib.pyplot as plt

print("Libraries matplotlib.pyplot imported successfully.")
```

2. Identify the numerical columns for univariate analysis and then generate their descriptive statistics, histograms, and box plots to visualize their distributions and identify outliers.



```python
numerical_cols = ['age', 'fare', 'sibsp', 'parch']

print("\n--- Univariate Analysis for Numerical Columns ---")

for col in numerical_cols:
    print(f"\nDescriptive Statistics for '{col}':")
    print(df[[col]].describe())

    # Create a figure with two subplots (side-by-side) for histogram and box plot
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Histogram
    sns.histplot(df[col].dropna(), kde=True, ax=axes[0])
    axes[0].set_title(f'Distribution of {col.capitalize()}')
    axes[0].set_xlabel(col.capitalize())
    axes[0].set_ylabel('Frequency')
    axes[0].grid(axis='y', linestyle='--', alpha=0.7)

    # Box Plot
    sns.boxplot(y=df[col], ax=axes[1])
    axes[1].set_title(f'Box Plot of {col.capitalize()}')
    axes[1].set_ylabel(col.capitalize())
    axes[1].grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

3. Identify the categorical and ordinal columns and then perform univariate analysis by printing their value counts and normalized percentages, and creating count plots to visualize their distributions.



```python
categorical_ordinal_cols = ['survived', 'pclass', 'sex', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone']

print("\n--- Univariate Analysis for Categorical/Ordinal Columns (Proportions) ---")

# Calculate the number of rows needed for a 3-column grid
num_cols_per_row = 3
num_rows = (len(categorical_ordinal_cols) + num_cols_per_row - 1) // num_cols_per_row

plt.figure(figsize=(num_cols_per_row * 5, num_rows * 4))

for i, col in enumerate(categorical_ordinal_cols):
    print(f"\nValue Counts for '{col}':")
    print(df[col].value_counts())
    print(f"\nNormalized Value Counts for '{col}' (Proportions):")
    norm_counts = df[col].value_counts(normalize=True).round(2)
    print(norm_counts)

    plt.subplot(num_rows, num_cols_per_row, i + 1)
    # Use barplot for proportions, mapping index (categories) to x and values (proportions) to y
    sns.barplot(x=norm_counts.index, y=norm_counts.values, palette='viridis', hue=norm_counts.index, legend=False)
    plt.title(f'Proportion of {col.capitalize()}')
    plt.xlabel(col.capitalize())
    plt.ylabel('Proportion')
    plt.ylim(0, 1) # Ensure y-axis is between 0 and 1 for proportions
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## Bivariate Analysis

Explore relationships between pairs of features using scatter plots, correlation matrices, and grouped visualizations to identify patterns and dependencies.


1. Calculate the correlation matrix for the specified numerical columns, including the target variable 'survived', and then visualize it using a heatmap to identify linear relationships between these variables.



```python
numerical_cols_for_corr = ['age', 'fare', 'sibsp', 'parch', 'survived']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols_for_corr].corr()

print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features and Survival')
plt.show()
```

2. Create a scatter plot to visualize the relationship between 'age' and 'fare', using 'survived' status to color the points, which will help in identifying patterns related to survival.



```python
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='age', y='fare', hue='survived', palette='viridis', alpha=0.7)
plt.title('Age vs. Fare, Colored by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='best')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

3. Explore relationships between numerical and categorical features, I will create box plots comparing the distribution of 'age' and 'fare' across 'survived', 'sex', and 'pclass', which will help visualize how these numerical features vary based on different categorical groups.



```python
numerical_vars = ['age', 'fare']
categorical_vars = ['survived', 'sex', 'pclass']

print("\n--- Bivariate Analysis: Numerical vs. Categorical Variables (Box Plots) ---")

# Define the grid dimensions for the subplots
num_rows = len(numerical_vars) # 2 rows for age and fare
num_cols = len(categorical_vars) # 3 columns for survived, sex, pclass

fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 6, num_rows * 5))
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

plot_index = 0
for num_col in numerical_vars:
    for cat_col in categorical_vars:
        ax = axes[plot_index]
        sns.boxplot(data=df, x=cat_col, y=num_col, hue=cat_col, palette='viridis', legend=False, ax=ax)
        ax.set_title(f'{num_col.capitalize()} by {cat_col.capitalize()}')
        ax.set_xlabel(cat_col.capitalize())
        ax.set_ylabel(num_col.capitalize())
        ax.grid(axis='y', linestyle='--', alpha=0.7)
        plot_index += 1

plt.tight_layout()
plt.show()
```

**Reasoning**:
To complete the bivariate analysis, I will now generate grouped bar charts to visualize the relationship between key categorical variables like 'survived' and 'sex', 'survived' and 'pclass', and 'survived' and 'embarked'.



```python
categorical_pairs = [
    ('survived', 'sex'),
    ('survived', 'pclass'),
    ('survived', 'embarked')
]

print("\n--- Bivariate Analysis: Categorical vs. Categorical Variables (Grouped Bar Charts - Proportions) ---")

# Define the grid dimensions for the subplots
num_rows = 2
num_cols = 3

# Create a figure and a set of subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 5, num_rows * 5))
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, (cat1, cat2) in enumerate(categorical_pairs):
    if i < len(axes): # Ensure we don't try to plot beyond the available subplots
        # Calculate proportions
        proportions = df.groupby(cat1)[cat2].value_counts(normalize=True).mul(100).rename('percentage').reset_index()

        sns.barplot(data=proportions, x=cat1, y='percentage', hue=cat2, palette='viridis', ax=axes[i])
        axes[i].set_title(f'Survival Proportion by {cat1.capitalize()} and {cat2.capitalize()}')
        axes[i].set_xlabel(cat1.capitalize())
        axes[i].set_ylabel('Percentage')
        axes[i].legend(title=cat2.capitalize())
        axes[i].grid(axis='y', linestyle='--', alpha=0.7)
        axes[i].tick_params(axis='x', rotation=0)
        axes[i].set_ylim(0, 100) # Set y-limit for percentages

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Missing Values and Outliers

Identify missing data and potential outliers, and discuss strategies for handling them during EDA.


1. Calculate and print the count of missing values for each column and then visualize these patterns using a heatmap to provide a clear overview of missing data distribution across the DataFrame.



```python
print("\n--- Missing Values Analysis ---")

# 1. Identify and count the number of missing values for each column
missing_values = df.isnull().sum()
print("\nMissing values count per column:")
print(missing_values[missing_values > 0])

# 2. Visualize the pattern of missing values using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()

# Percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nPercentage of missing values per column:")
print(missing_percentage[missing_percentage > 0].sort_values(ascending=False))
```

### Strategies for Handling Missing Values

Based on the analysis of missing values, here are strategies for handling the identified columns:

*   **`deck` (77.22% missing)**:
    *   **Strategy**: Given that `deck` has a very high percentage of missing values (over 77%), it is generally advisable to **drop this column**. Imputing such a large portion of missing data could introduce significant noise and bias, making the imputed values unreliable. Alternatively, if `deck` is considered potentially useful (e.g., if we extract the first letter for the deck level), we could create a new feature for 'deck level' and then drop the original `deck` column, allowing its missing values to be treated as a separate category (e.g., 'Unknown'). However, for simplicity and high missingness, dropping is often the most practical first step in EDA.

*   **`age` (19.87% missing)**:
    *   **Strategy**: `age` is a numerical variable with a moderate percentage of missing values. Common imputation strategies include:
        *   **Mean/Median Imputation**: Replacing missing values with the mean or median age. The median is often preferred for skewed distributions or in the presence of outliers. Given the box plot for `age`, which shows some outliers but a relatively symmetrical distribution around the median, both mean and median are viable. For robust imputation, the **median** is a good choice.
        *   **Regression Imputation**: Predicting `age` based on other related features (e.g., `pclass`, `sex`). This is a more sophisticated method but might be overkill for initial EDA.
        *   **Using `who` or `pclass` specific median/mean**: We could impute age based on the median age of specific groups (e.g., median age for men in 1st class). This can be more accurate than a global median.
    *   **Decision for EDA**: For initial EDA, **median imputation** is a robust and straightforward approach for `age`.

*   **`embarked` and `embark_town` (0.22% missing each)**:
    *   **Strategy**: Both `embarked` and `embark_town` are categorical variables with a very small number of missing values (only 2 out of 891 entries). For such a small proportion, the most common strategies are:
        *   **Mode Imputation**: Replacing missing values with the most frequent category (the mode). This is a safe approach when the number of missing values is negligible.
        *   **Deletion**: Removing the rows with missing values. Since there are only 2 rows, this would have a minimal impact on the overall dataset size.
    *   **Decision for EDA**: **Mode imputation** is generally preferred over deletion to retain all data points, especially if the missingness isn't random.

**Summary of Missing Value Handling Plan:**
1.  **Drop** the `deck` column due to excessive missingness.
2.  **Impute** missing `age` values with the **median age** per group.
3.  **Impute** missing `embarked` and `embark_town` values with their respective **modes**.

### Strategies for Handling Outliers

Revisiting the box plots generated for numerical variables (`age`, `fare`, `sibsp`, `parch`), we can observe the following regarding outliers:

*   **`age`**: The box plot for `age` shows a few individual points beyond the upper whisker, indicating some outliers (older passengers). These are generally valid data points (e.g., very old passengers) and might not necessarily be errors. However, they could influence some models.
    *   **Strategy**: For `age`, given that these are likely valid data, the primary approach for EDA is to **understand their context**. If further modeling requires outlier treatment, **winsorization (capping)** could be an option to limit extreme values without removing them entirely. Alternatively, some models are naturally robust to outliers.

*   **`fare`**: The `fare` box plot shows a significant number of extreme outliers, particularly on the higher end. This is expected as some passengers paid very high fares for first-class travel or luxury accommodations.
    *   **Strategy**: `fare` is highly skewed and contains many extreme values. For EDA, it's crucial to acknowledge these and understand their impact. For modeling, **log transformation** is a common technique to reduce the skewness and mitigate the impact of extreme values, making the distribution more normal-like. Another option could be **capping** at a certain percentile (e.g., 99th percentile) if the extreme values are deemed overly influential for certain analyses.

*   **`sibsp` and `parch`**: Both `sibsp` (siblings/spouses aboard) and `parch` (parents/children aboard) show outliers, indicating a small number of passengers traveling with a large family contingent (e.g., 5 or 8 siblings/spouses, or 4-6 parents/children). These are likely valid but rare occurrences.
    *   **Strategy**: Similar to `age`, these are likely valid data points. For `sibsp` and `parch`, it's generally best to **keep these values as they are**, as they represent real family sizes and could be highly predictive of survival (e.g., families might stick together). If models are highly sensitive, they could be **capped** at a reasonable upper bound (e.g., top 1% or 5%) or **binned** into categories (e.g., 0, 1-2, 3+ family members) to reduce the impact of extreme counts.

**Summary of Outlier Handling Plan:**
1.  **`age`**: Understand context; potentially winsorize if needed for specific models, but largely retain.
2.  **`fare`**: Consider **log transformation** to reduce skewness and impact of extreme values, or capping at a high percentile.
3.  **`sibsp`, `parch`**: Retain as they are likely valid and informative; potentially bin or cap if highly influential for specific analyses.
